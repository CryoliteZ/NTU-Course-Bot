```python
newdf = pd.read_csv('data/ntucourse.csv', delimiter=';', index_col=[0,1,2])


def parseEN(line):
    line = re.sub("[^A-Za-z]", "", line.strip())
    return line
newdf['CDEP'] = newdf['CONUM'].apply(parseEN)
newdf.head(2)
newdf.to_csv('data/ntucourse.csv',sep=';',  float_format= '%.3f')
```
